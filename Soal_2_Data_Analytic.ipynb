2.3. Presentase pria dan wanita

```python
gender_labels = ['Pria' , 'Wanita']
pria = len(profesi[profesi['gender'] == 'M'])
wanita = len(profesi[profesi['gender'] == 'F'])
jumlah_gender = [pria,wanita]


# fig = plt.figure(figsize =(10, 7))
#plt.pie(jumlah_gender, labels = gender_labels)

# Creating explode data
explode = (0.1, 0.0)

# Creating color parameters
colors = ( "orange", "blue")

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(jumlah_gender,
                                  autopct = lambda pct: func(pct, jumlah_gender),
                                  explode = explode,
                                  labels = gender_labels,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))

# Adding legend
ax.legend(wedges, gender_labels,
          title ="Gender",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Presentase pria dan wanita")

plt.show()
```
